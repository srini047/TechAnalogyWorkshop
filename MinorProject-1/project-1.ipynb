# Implement Simple Linear Regression

## Importing Necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Reading the Data

```python
df = pd.read_csv('salary_data.csv')
df
```

## Getting Info About DataFrame

### Shape

```python
df.shape
```

### Information

```python
df.info()
```

## Data Modelling

### Splitting Labels and Predictors

```python
X = df.iloc[:, :-1]
y = df.iloc[:, :1]
```

### Splitting the Train and Test Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

### Build a LinearRegression Model

```python
model = LinearRegression()
```

### Fit the Model with the Train Dataset

```python
model.fit(X_train, y_train)
```

## Data Visualization

### Before Training

```python
plt.plot(df['YearsExperience'], df['Salary'], marker = 'o', color = 'green')
plt.title('Years VS Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Training Visualisation

```python
plt.scatter(X_train, y_train, color = 'green', marker = '+')
plt.plot(X_train, model.predict(X_train), color = 'orange')
plt.title('Training Dataset')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Testing Visualisation

```python
plt.scatter(X_test, y_test, color = 'blue', marker = '+')
plt.plot(X_test, model.predict(X_test), marker = '+', color = 'red')
plt.title('Testing Dataset')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Model Prediction

```python
y_pred = model.predict(X_test)
y_pred
```
